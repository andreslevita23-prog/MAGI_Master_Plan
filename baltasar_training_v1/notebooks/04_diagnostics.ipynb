```python
target_audit = pd.read_csv(metrics_dir / f'{run_id}_target_audit.csv')
target_audit
```

```python
class_metrics = pd.read_csv(metrics_dir / f"{run_id}_{summary['best_model_name']}_class_metrics.csv")
class_metrics
```

```python
walk_forward = pd.read_csv(metrics_dir / f'{run_id}_walk_forward_metrics.csv')
walk_forward
```

```python
report_path = reports_dir / f'{run_id}_diagnostic_report.md'
print(report_path.read_text(encoding='utf-8'))
```
